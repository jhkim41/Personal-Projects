```python
firm_df= pd.read_excel('/Users/jonghyunkim/repo/Tristan/S&P 500 Companies (Standard and Poor 500).xlsx')
firm_df
```

```python
driver = webdriver.Chrome('/Users/jonghyunkim/repo/Tristan/chromedriver')
```

```python
driver.maximize_window()
# url = 'https://www.fool.com/'
url = 'https://www.fool.com/quote/nasdaq/aapl/#quote-earnings-transcripts'
driver.get(url)
time.sleep(3)
```

```python
symbol = []
name = []
web_tickers = []
years = []
quarters = []
target_texts = []
esg_word_count = []
sustainability_word_count = []
res_man_word_count = []

checking = pd.read_excel('SeokJun Collection_체크포인트용.xlsx')
checking2 = pd.read_excel('SeokJun Collection_체크포인트용2.xlsx')
checking3 = pd.read_excel('SeokJun Collection_체크포인트용3.xlsx')
checking4 = pd.read_excel('SeokJun Collection_체크포인트용4.xlsx')
checking5 = pd.read_excel('SeokJun Collection_체크포인트용5.xlsx')
checking6 = pd.read_excel('SeokJun Collection_체크포인트용6.xlsx')


for index, row in firm_df.iterrows():

    if row['Symbol'] in list(checking['Symbol']):
        continue
    elif row['Symbol'] in list(checking2['Symbol']):
        continue

    elif row['Symbol'] in list(checking3['Symbol']):
        continue
    elif row['Symbol'] in list(checking4['Symbol']):
        continue

    elif row['Symbol'] in list(checking5['Symbol']):
        continue
    elif row['Symbol'] in list(checking6['Symbol']):
        continue

    try:
        ## 검색창에 원하는 검색어 입력 후 대기
        driver.find_element(By.XPATH, '//*[@id="tophat"]/div[1]/input').send_keys(row['Name'] + ' ')
        time.sleep(5)

        # 검색창에 엔터 클릭
        driver.find_element(By.XPATH, '//*[@id="tophat"]/div[1]/input').send_keys(Keys.ENTER)
        time.sleep(4)

        Title = driver.find_element(By.XPATH, '/html/body/div[9]/div[3]/div[1]/section[1]/div[1]/header/div/div[2]/h1').text
        web_ticker = Title.split(': ')[-1][:-1]
        if web_ticker != row['Symbol']:
            driver.find_element(By.XPATH, '//*[@id="tophat"]/div[1]/input').send_keys(row['Symbol'] + ' ')
            time.sleep(5)
            driver.find_element(By.XPATH, '//*[@id="tophat"]/div[1]/input').send_keys(Keys.ENTER)
            time.sleep(4)

            Title = driver.find_element(By.XPATH, '/html/body/div[9]/div[3]/div[1]/section[1]/div[1]/header/div/div[2]/h1').text
            web_ticker = Title.split(': ')[-1][:-1]
    except:
        symbol.append(row['Symbol'])
        name.append(row['Name'])
        web_tickers.append(web_ticker)
        years.append('error')
        quarters.append('error')
        target_texts.append('error')
        esg_word_count.append('error')
        sustainability_word_count.append('error')
        res_man_word_count.append('error')
        continue

    time.sleep(2) # 페이지 로딩 대기

    try:
        ## 페이지 좌측 메뉴 바에 Earnings Transcript 라는 버튼 클릭 후 로딩 대기

        # driver.find_element(By.XPATH, '/html/body/div[9]/div[3]/div[1]/section[2]/div/div[1]/div[1]/div').find_element(By.XPATH, "//button[text()='Earnings Transcripts']").click()
        driver.find_element(By.XPATH, '//*[@id="quote-nav"]/div/div[2]/nav/ul/li[8]/a').click()
        time.sleep(2)

        ## 가끔 스크롤 더 혹은 덜 내리는 경우가 잇으니 다시한번 클릭
        driver.find_element(By.CSS_SELECTOR, 'body').send_keys(Keys.PAGE_UP)
        driver.find_element(By.CSS_SELECTOR, 'body').send_keys(Keys.PAGE_UP)
        driver.find_element(By.XPATH, '//*[@id="quote-nav"]/div/div[2]/nav/ul/li[8]/a').click()

        time.sleep(2)

        ## View more earnings transcript 버튼 누르기 2번
        driver.find_element(By.XPATH, '//*[@id="quote-earnings-transcripts"]/button').click()
        time.sleep(2)
        driver.find_element(By.XPATH, '//*[@id="quote-earnings-transcripts"]/button').click()
        time.sleep(2)
    except:
        symbol.append(row['Symbol'])
        name.append(row['Name'])
        web_tickers.append(web_ticker)
        years.append('error')
        quarters.append('error')
        target_texts.append('error')
        esg_word_count.append('error')
        sustainability_word_count.append('error')
        res_man_word_count.append('error')

        continue


    ## 각 earnings transcript의 path 가져오기
    path_list = driver.find_element(By.ID, 'quote-earnings-transcripts').find_elements(By.CLASS_NAME, 'mb-6')

    ## 각 path의 제목들 가져오기 (언제건지 확인용)
    candidate_list = [path_list[n].text for n in range(len(path_list))]


    ## 2023년과 2021년에 해당하는 애들 가져오기
    i = 0
    target_indexs = []
    for c in candidate_list:
        if '2023' in c:
            target_indexs.append(i)
        elif '2021' in c:
            target_indexs.append(i)
        i += 1
    
    for k in target_indexs:
        try:
            path_list[k].click()
        except:
            driver.find_element(By.CSS_SELECTOR, 'body').send_keys(Keys.PAGE_DOWN)
            path_list[k].click()
        time.sleep(2)
        
        symbol.append(row['Symbol'])
        name.append(row['Name'])
        web_tickers.append(web_ticker)

        ## 위에서 가져온 path 제목에서 quarter와 year 골라내서 추가하기
        title_texts = candidate_list[k].split('(' + web_ticker + ')')[1].split()
        quarter = title_texts[0]
        year = title_texts[1]

        years.append(year)
        quarters.append(quarter)
        
        ## 텍스트 가져와서 소문자화 등 전처리 한 후 고유 단어 개수세기
        target_text = driver.find_element(By.CLASS_NAME, 'article-body').text
        target_text = target_text.replace('\n', ' ')
        target_texts.append(target_text)

        target_text = target_text.lower()
        
        vectorizer = CountVectorizer(ngram_range=(1,2))
        vec_reviews = vectorizer.fit_transform([target_text])
        vec_terms = vectorizer.get_feature_names_out()
        vec_df = pd.DataFrame(vec_reviews.toarray(), columns=vec_terms)

        if 'esg' in vec_df.columns:
            esg_count = vec_df['esg']
        else:
            esg_count = 0

        esg_word_count.append(esg_count)

        if 'sustainability' in vec_df.columns:
            sus_count = vec_df['sustainability']
        else:
            sus_count = 0
        
        sustainability_word_count.append(sus_count)


        if 'responsible management' in vec_df.columns:
            res_man_count = vec_df['responsible management']
        else:
            res_man_count = 0
        
        res_man_word_count.append(res_man_count)

        driver.back()
        time.sleep(2)
    
    final_data = pd.DataFrame({'Symbol' : symbol, 'Name' : name, 'Website_ticker' : web_tickers, 'Year' : years, 'Quarter' : quarters, 'Raw Text' : target_texts, 'ESG' : esg_word_count, 'Sustainability' : sustainability_word_count, 'Responsible Management' : res_man_word_count})
    final_data.to_excel('SeokJun Collection_체크포인트용7.xlsx')
```
